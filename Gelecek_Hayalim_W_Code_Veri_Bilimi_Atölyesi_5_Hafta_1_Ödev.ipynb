EDA (Keşifsel Veri Analizi)

```python
plt.figure(figsize=(8, 4))
sns.countplot(x='Mental_Health_Condition', data=df)
plt.title('Zihinsel Sağlık Durumu Dağılımı')
plt.xlabel('Zihinsel Sağlık Durumu')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.histplot(df['Age'], bins=15)
plt.title('Yaş Dağılımı')
plt.xlabel('Yaş')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.boxplot(x='Job_Role', y='Stress_Level', data=df)
plt.title('İş Rolüne Göre Stres Seviyesi')
plt.xlabel('İş Rolü')
plt.ylabel('Stres Seviyesi')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Gender', y='Work_Life_Balance_Rating', data=df)
plt.title('Cinsiyete Göre İş-Yaşam Dengesi Puanları')
plt.xlabel('Cinsiyet')
plt.ylabel('İş-Yaşam Dengesi Puanı')
plt.show()
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import io
```

```python
from google.colab import files
uploaded = files.upload()
```

```python
df = pd.read_csv(io.BytesIO(uploaded['gym_members_exercise_tracking (2).csv']))
```

```python
print(df.head())
```

```python
print("Veri Setinin Boyutu:", df.shape)
```

```python
print("Veri Türleri ve Eksik Değerler:")
print(df.info())
```

```python
gender_counts = df['Gender'].value_counts()

plt.figure(figsize=(5, 4))
gender_counts.plot(kind='bar', color='plum')
plt.title('Cinsiyete Göre Katılımcı Sayısı')
plt.xlabel('Cinsiyet')
plt.ylabel('Katılımcı Sayısı')
plt.show()
```

```python
workout_counts = df['Workout_Type'].value_counts()
colors = ['lightblue', 'lightgreen', 'salmon', 'plum', 'lightcoral']
plt.figure(figsize=(5, 5))
plt.pie(workout_counts, labels=workout_counts.index, autopct='%1.1f%%', startangle=140, colors=colors)
plt.title('Çalışma Türlerine Göre Katılımcı Dağılımı')
plt.axis('equal')
plt.show()
```

```python
plt.figure(figsize=(8, 5))
plt.hist(df['Age'], bins=15, color='plum', edgecolor='red')
plt.title('Yaş Dağılımı')
plt.xlabel('Yaş')
plt.ylabel('Katılımcı Sayısı')
plt.grid(axis='y')
plt.show()
```

```python
plt.figure(figsize=(7, 5))
plt.scatter(df['Session_Duration (hours)'], df['Calories_Burned'],  color='orchid')
plt.title('Çalışma Süresi ile Kalori Tüketimi Arasındaki İlişki')
plt.xlabel('Çalışma Süresi (saat)')
plt.ylabel('Yakılan Kalori')
plt.grid()
plt.show()
```

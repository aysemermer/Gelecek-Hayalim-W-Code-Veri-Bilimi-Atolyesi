```python
countries = df['Country'].unique()
print(f"Veri setinde toplam {len(countries)} farklı ülke var.")
```

```python
print("Ülkeler:")
for country in countries:
    print(country)
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Gender Inequality Index (2021)', y='Country', data=df.sort_values(by='Gender Inequality Index (2021)', ascending=False).head(10)) #“ascending = false” → azalan şekilde sıralama yapılmasını sağlar.
plt.title("2021 Yılında En Yüksek Cinsiyet Eşitsizliği Olan 10 Ülke")
plt.xlabel("Cinsiyet Eşitsizliği Endeksi (2021)")
plt.ylabel("Ülkeler")
plt.show()
```

```python
plt.figure(figsize=(8, 4))
sns.barplot(x='Gender Inequality Index (2021)', y='Country', data=df.sort_values(by='Gender Inequality Index (2021)').head(10))
plt.title("2021 Yılında En Düşük Cinsiyet Eşitsizliği Olan 10 Ülke")
plt.xlabel("Cinsiyet Eşitsizliği Endeksi (2021)")
plt.ylabel("Ülkeler")
plt.show()
```

```python
plt.figure(figsize=(7, 4))
sns.countplot(y='Human Development Groups', data=df, order=df['Human Development Groups'].value_counts().index)
plt.title("İnsan Gelişimi Gruplarına Göre Ülke Sayısı")
plt.xlabel("Ülke Sayısı")
plt.ylabel("İnsan Gelişimi Grupları")
plt.show()
```

```python
ulke1 = 'Turkey'
ulke1_verisi = df[df['Country'] == ulke1]
yillar = ['Gender Inequality Index (' + str(year) + ')' for year in range(1990, 2022)]
gii_degerleri = ulke1_verisi[yillar].values.flatten()
plt.figure(figsize=(10, 5))
plt.plot(yillar, gii_degerleri, marker='o')
plt.title(ulke1 + " Cinsiyet Eşitsizliği Endeksi (1990-2021)")
plt.xlabel("Yıllar")
plt.ylabel("Cinsiyet Eşitsizliği Endeksi (GII)")
plt.grid()
plt.xticks(rotation=45)
plt.show()
```
